```python
import pandas as pd
import numpy as np
```

```python
Demography=pd.read_csv("Demographic.csv")
Demography.head()
```

```python
MonthlySales=pd.read_csv("Monthly Sales.csv")
MonthlySales.head()
```

```python
Cities=MonthlySales['CITY'].unique()
Cities=Cities.tolist()
print (Cities)
```

```python
#Selecting the Cities in question from the Demographic data
Filter= Demography['City'].isin(Cities)
Demography['City'].where(Filter, inplace=True)
Demography.dropna(inplace=True)
Demography.to_csv("demographyOfStores.csv") #The Cleaned Demographic Data
```

```python
Census_Data=pd.read_csv("Web Scraped.csv")
Census_Data.head()
```

```python
#df[['First','Last']] = df.Name.str.split("_",expand=True,)
Cities=['Buffalo', 'Casper', 'Cheyenne', 'Cody', 'Douglas', 'Evanston', 'Gillette', 'Powell', 'Riverton', 'RockSprings', 'Sheridan']
```

```python
Census_Data[['City','County']]=Census_Data["City|County"].str.split("|",expand=True)
Census_Data['City']=Census_Data['City'].str.replace("?","")
Census_Data['City']=Census_Data['City'].str.replace(" ?","")

Filter= Census_Data['City'].isin(Cities)
Census_Data['City'].where(Filter, inplace=True)
Census_Data.dropna(inplace=True)
Census_Data['City']=Census_Data['City'].str.replace("RockSprings","Rock Springs")
Census_Data
```

```python
Census_Data=Census_Data.drop(['City|County','County','2014 Estimate','2000 Census'],1)
Census_Data
```

```python
Census_Data['2010 Census']=Census_Data['2010 Census'].str.replace("<td>","")
Census_Data['2010 Census']=Census_Data['2010 Census'].str.replace("</td>","")
Census_Data['2010 Census']=Census_Data['2010 Census'].str.replace(",","")
Census_Data['2010 Census']=Census_Data['2010 Census'].astype("int")
Census_Data.to_csv("CensusDataCleaned.csv")
```
